```python
def find_not_and_combine(x):
    pattern = '(not)\s(\w+)'
    temp = re.split(pattern,x)
    new = temp.copy()
    remove_elements = []
    if len(temp) > 1:
        for i in range(len(temp)-1):
            if temp[i] == 'not':
                remove_elements.append(temp[i+1])
                new[i] = temp[i]+temp[i+1]
        for j in remove_elements:
            if j in new: 
                new.remove(j)
    final_list = []
    for i in new:
        for j in i.split():
            final_list.append(j)
    final_results = ' '.join(final_list)
    return final_results
```

```python
new_review_im['combine_not'] = new_review_im['no_digits'].map(lambda x: find_not_and_combine(x))
```

### Remove punctuation and stop words

```python
import string
from string import digits 
```

```python
# remove punctuation with white space
new_review_im['no_punctuation'] = new_review_im['remove_name'].map(lambda x: x.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation))))
```

```python
# tokennize each review: generate list of words
new_review_im['words'] = new_review_im['no_punctuation'].map(lambda x: word_tokenize(x))
```

```python
from nltk.corpus import stopwords
from wordcloud import STOPWORDS
stopwords1 = list(STOPWORDS) 
stopwords2 = stopwords.words('English')
```
