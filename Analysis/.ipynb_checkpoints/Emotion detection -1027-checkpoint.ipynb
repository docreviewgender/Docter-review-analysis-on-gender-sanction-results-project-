## Further split each sentence into a list of words

```python
from nltk.tokenize import word_tokenize
```

```python
import string
from string import digits 
```

```python
def remove_punctuation(sentences):
    new_sentences=[]
    for i in sentences:
        i = i.translate(str.maketrans('', '', string.punctuation))
        new_sentences.append(i)
    return new_sentences
```

```python
# remove punctuation within a review
review['remove_punc'] = review['sentence'].map(lambda x: remove_punctuation(x))
```

```python
review['word'] = review['remove_punc'].map(lambda sentences: [word_tokenize(sentence) for sentence in sentences])
```

### Read NRC-Emotional Lexicon

```python
list_e = [line.split('\t') for line in open('/Users/chongchen/Desktop/19Fall RA/NRC-Sentiment-Emotion-Lexicons/NRC-Sentiment-Emotion-Lexicons/NRC-Emotion-Lexicon-v0.92/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt')]
```

#### Generate the nrc emotional lexicon dictionary

```python
nrc_dict = {}
for i in list_e:
    if i[0] not in nrc_dict.keys():
               nrc_dict[i[0]] = {i[1]:int(i[2].strip('\n'))}
    elif i[0] in nrc_dict.keys():
               nrc_dict[i[0]][i[1]] = int(i[2].strip('\n'))
```

```python
nrc_dict
```

```python
anticipation: [['friendly'], ['time', 'glad'], ['time']]
joy: [['clean', 'friendly'], ['love', 'glad'], ['helpful']]
positive: [['clean', 'friendly'], ['love', 'doctor', 'glad', 'doctor'], ['helpful', 'solution']]
trust:  [['clean', 'friendly'], ['doctor', 'doctor'], ['helpful']]
```

```python
def emotion_score(sentences,emotion):
    score_list = []
    for sentence in sentences:
        score = 0
        num_of_words = len(sentence)
        for word in sentence:
            if nrc_dict.get(word, 0) != 0:
                s = nrc_dict.get(word, 0).get(emotion,0)
            else: 
                s = 0
            score = score+s
        if num_of_words != 0:
            final_score = score/num_of_words
        else:
            final_score = 0
        score_list.append(final_score)
    avg_score = sum(score_list)/len(score_list)
    
    return avg_score
```

```python
emotion_score(review['word'][0],'anticipation')
```

```python
def get_words(sentences, emotion):
    word_list = []
    for sentence in sentences:
        words = []
        for word in sentence:
            if nrc_dict.get(word, 0) != 0:
                s = nrc_dict.get(word, 0).get(emotion,0)
                if s != 0:
                    words.append(word)
        word_list.append(words)
    return word_list
```

```python
emotion_list = ['anger','anticipation','disgust','fear','joy','negative','positive','sadness','surprise','trust']
for i in emotion_list:
    review[i] = review['word'].map(lambda x: emotion_score(x,i))
    review[i+' words'] = review['word'].map(lambda x: get_words(x,i))
```

```python
name_dict = {}
for i in emotion_list:
    name_dict[i+'words']= i+'_words'
```

```python
review.rename(columns = name_dict, inplace = True)
```

Calculate average lenth of reviews

```python
def avg_length(sentences):
    num_sentence = len(sentences)
    a = 0
    for sentence in sentences:
        a = a + len(sentence)
    avg_length = a/num_sentence
    return avg_length
```

```python
review['avg_length'] = review['word'].map(lambda x: avg_length(x))
```
