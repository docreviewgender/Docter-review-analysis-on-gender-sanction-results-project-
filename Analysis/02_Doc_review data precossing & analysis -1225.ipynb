```python
# Generate a new dataframe of doc's review, make sure each record has single review
review_im = pd.DataFrame()
for i, sentence in enumerate(doc_im['sentence']):
    temp_dict = dict(enumerate(sentence))
    a = len(list(temp_dict.keys()))
    s = str(doc_im['hp_id'][i])
    temp_df = pd.DataFrame.from_dict(data = temp_dict, orient = 'index', columns=['Review'])
    temp_df['hp_id'] = [s for i in range(a)]
    review_im = review_im.append(temp_df, ignore_index=True)
```

### 2. Pre-processing review: lower case and remove punctuation, stopwords

```python
# lower case all the reviews
review_im['lower'] = review_im['Review'].map(lambda x: x.lower())
```

```python
import string
from string import digits 
```

```python
# using translate and digits 
# to remove numeric digits from string 
remove_digits = str.maketrans('', '', digits) 
review_im['no_digits'] = review_im['lower'].map(lambda x: x.translate(remove_digits))
```

```python
# remove punctuation
# review_im['no_punctuation'] = review_im['no_digits'].map(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
```

```python
# remove punctuation with white space
review_im['no_punctuation'] = review_im['no_digits'].map(lambda x: x.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation))))
```

```python
print(review_im['no_punctuation'][0])
```

```python
from nltk.tokenize import word_tokenize
```

```python
# tokennize each review: generate list of words
review_im['words'] = review_im['no_punctuation'].map(lambda x: word_tokenize(x))
```

```python
from nltk.corpus import stopwords
from wordcloud import STOPWORDS
stopwords1 = list(STOPWORDS) 
stopwords2 = stopwords.words('English')
```

```python
## combine stopwords1 & stopwords2
for i in stopwords2:
    if i not in stopwords1:
        stopwords1.append(i)
stopwords1.append('')
stopwords1.append(' ')
```

```python
# remove stop words
review_im['words']=review_im['words'].map(lambda x: [i for i in x if i not in stopwords1])
```

### 3. Sampling equal numeber of reviews of the two genders, tag documents.

```python
review_im['gender'].value_counts()
```

```python
review = list(review_im['words'])
```

```python
review[0][0]
```

```python
gender = list(review_im['gender'])
```

```python
len(review),len(gender)
```

```python
# Tag document
import gensim
docs = [] # each review/row is a tagged document.
for i in range(len(review)):
    docs.append(gensim.models.doc2vec.TaggedDocument(review[i], [gender[i]]))
```

```python
docs[1]
```

```python
docs_for_vocab = [] 
for i in docs:
    docs_for_vocab.append(i)
```

```python
import random
```

```python
repeat = {}
indices = []
batch = []
draw = 10000
m_candidates = list(review_im[review_im['gender']=='M'].index)
f_candidates = list(review_im[review_im['gender']=='F'].index)
for b in range(draw):
    m = random.choice(m_candidates)
    indices.append(m)
    f = random.choice(f_candidates)
    indices.append(f)
for i in indices:
    try:
        repeat[i]+=1
    except:
        repeat[i]=1
    batch.append(docs[i])
```

### 4. Import Gensim, run doc2vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import multiprocessing
```

```python
cores = multiprocessing.cpu_count()
```

```python
model = Doc2Vec(dm=0, window=20,vector_size=300,min_count=30,epochs=5,workers=cores,
               hs=0,negative=7,dbow_words=1,dm_concat=1)
model.build_vocab(docs_for_vocab)
model.train(documents = batch,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
stopwords = set(STOPWORDS) 
stopwords.add('amp')
stopwords.add('ive')
stopwords.add('im')
```

```python
corpus=female_words, male_words
cv=CountVectorizer(stop_words=stopwords)
word_count_vector=cv.fit_transform(corpus)
```

```python
#reverse key-value pair in dict(cv.vocabulary)
cv_voc_reverse={int(value):key for key,value in cv.vocabulary_.items()}

wordcount_Matrix=pd.DataFrame(word_count_vector.toarray())
```

```python
wordcount_Matrix.rename(index={0:'female',1:'male'},columns=cv_voc_reverse,inplace=True)
```

```python
wordcount_Matrix.head()
```

```python
wordcount_Matrix['sum']=wordcount_Matrix.sum(axis=1)
wordcount_percent_Matrix=wordcount_Matrix.div(wordcount_Matrix['sum'],axis=0)
wordcount_percent_Matrix_trans=wordcount_percent_Matrix.transpose()
```

```python
# get the top 100 freaquent word
female_top500=wordcount_percent_Matrix_trans.nlargest(501,'female')['female']
male_top500=wordcount_percent_Matrix_trans.nlargest(501,'male')['male']
```

```python
gender_ratio_ser=pd.Series(gender_ratio)
gender_ratio_ser_recipro=pd.Series([x if x>=1 else 1/x for x in gender_ratio_ser],index=gender_ratio_ser.index)
gender_ratio_ser_gender=pd.Series(['M'if x>1 else 'F' for x in gender_ratio_ser],index=gender_ratio_ser.index)
gender_ratio_df=pd.concat([gender_ratio_ser,gender_ratio_ser_recipro,gender_ratio_ser_gender],axis=1)
gender_ratio_df.columns=['ratio','cv_ratio','gender']
```
