```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from nltk.tokenize import word_tokenize
```

```python
path = "/Users/chongchen/Desktop/19Fall RA/data/raw data/"
doc_review = pd.read_csv(path + 'docs_before.csv', index_col = 0)
gender = pd.read_csv(path + 'genders.csv').drop_duplicates()
doc_review = pd.merge(doc_review,gender,on='hp_id',how='inner')
```

```python
doc_review['hasorder'] = doc_review['hasorder'].map(lambda x: str(x))
```

```python
def no_space(x):
    a = x.split(' ')
    b = ''
    for i in a:
        if i != '':
            b = b+i
    return b
doc_review['spec_comb'] = doc_review['spec_comb'].map(lambda x:no_space(x))
```

Focus on Internal Medicine Doctors

```python
im_data = doc_review[doc_review['spec_comb']=='InternalMedicine']
```

```python
im_data
```

#### Convert the review corpus into a list of sentences

```python
im_data.loc[:,['sentence']]= im_data['review_corpus'].map(lambda x: x.split('|'))
```

```python
im_data['word'] = im_data['sentence'].map(lambda x: [word_tokenize(i) for i in x] )
```

#### Calculate number of reviews of each doctor

```python
im_data.loc[:,['Number of Review']] = im_data['sentence'].map(lambda x: len(x))
```

#### Calculate average length of reviews of each doctor

```python
def sum_words(x):
    words = 0
    for i in x:
        words = len(i)
    return words
```

```python
im_data['averaeg length'] = im_data['word'].map(lambda x: sum_words(x)/len(x))
```
